1) Crie um código que receba o nome do aluno, das notas e informa a média do aluno.

2) Crie um código que receba as medidas de um retângulo e informe o perimetro do mesmo.

```python

base = float(input("Digite o valor da base do retângulo: "))
altura = float(input("Digite o valor da altura do retângulo: "))

perimetro = 2 * (base + altura)


print(f"O perímetro do retângulo é: {perimetro}")
```
